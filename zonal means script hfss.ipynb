```python
##importing packages needed
from netCDF4 import Dataset
import netCDF4
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import xarray as xr
import cartopy.crs as crs
```

```python
### import file
file = "hfss_Amon_UKESM1-0-LL_ssp370_r1i1p1f2_gn_201501-204912.nc"
dataset = xr.open_dataset(file)
```

```python
## latstart = 70 #70N-90N,
latstart=70
latend = 90


varnear = dataset.hfss.sel(lat=slice(latstart, latend))
# Select the time period 
varnear_bp = varnear.sel(time=slice("2020", "2040"))
# Compute the mean along the time axis
varnearplot = varnear_bp.mean(dim="time")
levels = [-70,-65,-60,-55,-50,-45,-40,-35,-30,-25,-20,-15,-10,-5, 0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75]
# Define an axes object with a projection, levels
varnearplot.plot(levels=levels)
```

```python
varnearplot.mean() 
```
