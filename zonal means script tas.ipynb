```python
##importing packages needed
from netCDF4 import Dataset
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import xarray as xr
import cartopy.crs as crs
```

```python
### import file 
file = "tas_Amon_UKESM1-0-LL_ssp370_r1i1p1f2_gn_201501-204912.nc"
dataset = xr.open_dataset(file)
```

```python
## latstart = 70 #70N-90N,
latstart=70
latend = 90


varnear = dataset.tas.sel(lat=slice(latstart, latend))
# Select the time period 
varnear_bp = varnear.sel(time=slice("2020", "2040"))
# Compute the mean along the time axis
varnearplot = varnear_bp.mean(dim="time")
levels = [240, 241, 242, 243, 244, 245,246,247,248,249, 250,251,252,253,254,255,256,257,258,259,
          260,261,262,263,264,265,266,267,268,269,270,271,272,273,274,275,276,277,278,279,280,
          281, 282, 283,284, 285]

# Define an axes object with a projection, levels
varnearplot.plot(levels=levels)
```
